```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
df = pd.read_csv("wine.csv")
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(df.corr(),annot=True,vmin=-1,vmax=1,cmap='RdBu')
plt.show()
```

```python
df['Type'].value_counts()
```

```python
df2=df.iloc[:,1:]
df2
```

```python
# Converting data to numpy array
df_arr=df2.values
df_arr
```

```python
# Normalizing the numerical data 
df_norm=scale(df_arr)
df_norm
```

```python
# Applying PCA Fit Transform to dataset
pca=PCA(n_components=13)
```

```python
df_pca=pca.fit_transform(df_norm)
df_pca
```

```python
# PCA Components matrix or covariance Matrix
pca.components_
```

```python
# The amount of variance that each PCA has
var=pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

```python
# plot between PCA1 and PCA2 
x = df_pca[:,0:1]
y = df_pca[:,1:2]
#z = pca_values[:2:3]
plt.scatter(x,y)
```

```python
# Final Dataframe
final_df=pd.concat([df['Type'],pd.DataFrame(df_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
sns.scatterplot(data=final_df, x='PC1', y='PC2', hue='Type');
```

# Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# As we already have normalized data, create Dendrograms
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(df_norm,'complete'))
```

```python
# Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(df_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
df3=df.copy()
df3['clustersid']=hclusters.labels_
df3
```
